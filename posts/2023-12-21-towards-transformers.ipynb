### Basic Imports

```python
import tiktoken
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
```

```python
encoding.encode("Hello World! This is a simple notebook")
```

```python
encoding.decode([9906, 4435, 0, 1115])
```

```python
ser = {}
n =20
for i in range(n**2):
    ser[i] = encoding.decode([i])
```

```python
import pandas as pd
```

```python
pd.DataFrame(pd.Series(ser).values.reshape(n,n))
```
